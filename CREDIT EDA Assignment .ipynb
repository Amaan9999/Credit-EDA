### Data cleaning:

#### Missing value treatment:

```python
df["ORGANIZATION_TYPE"]
```

Some values in ORGANIZATION_TYPE has XNA values .So assuming these values to be null, we can replace them with Nan or null values


```python
df.replace("XNA",np.nan,inplace=True) #Replacing XNA values with null values
```

```python
df2.replace("XNA",np.nan,inplace=True)#Replacing XNA values with null values in df2
df2.replace("XAP",np.nan,inplace=True)#Replacicng XAP values with null values in df2
```

```python
df2["SELLERPLACE_AREA"].replace(-1,np.nan,inplace=True)#Replacing -1 values with null values in df2
```

```python
#calculating the percentage of null values in each column of df
df.isnull().sum()*100/len(df)
```

```python
#Dropping all columns which have null values greater than 40 % in df
n = (df.isnull().sum() *100)/ len(df)
df.drop(n[n>40].index, axis=1, inplace=True)
```

```python
#again  checking for total  number of null values in each column in df
df.isnull().sum()
```

```python
# dropping all documents columns in df1 since these are not relevant in our analysis
df.drop(['FLAG_DOCUMENT_2', 'FLAG_DOCUMENT_3',
       'FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5', 'FLAG_DOCUMENT_6',
       'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8', 'FLAG_DOCUMENT_9',
       'FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11', 'FLAG_DOCUMENT_12',
       'FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14', 'FLAG_DOCUMENT_15',
       'FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17', 'FLAG_DOCUMENT_18',
       'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20', 'FLAG_DOCUMENT_21'],axis=1,inplace=True)
```

```python
#dropping HOUR_APPR_PROCESS_START since this has no relevance in our analysis
df.drop(["HOUR_APPR_PROCESS_START"],axis=1,inplace=True)
```

```python
df.head()
```

```python
#  Now we can delete all the rows that have  only null values in all columns since these rows will not be of any use to us
df = df.dropna(how='all')
```

```python
#Doing the same for   df2 
df2 = df2.dropna(how='all')
```
